Some typical NumPy 'random' functions:

```python
import numpy as np

# To generate a random float values between 0 and 1 (exclusive)
print(np.random.rand(3, 2)) # Creates a 3 x 2 matrix of random float values between 0 and 1 (exclusive)

# To generate random float values from standard normal distribution (mean = 0 and std = 1)
print(np.random.randn(5)) # Generates 5 random float values from standard normal distribution

# To generate random integer values within a range of values
print(np.random.randint(1, 100, 10)) # Generates 10 random integer values between 1 and 100 (exclusive)

# To randomly select an element from a given list of elements
print(np.random.choice([1, 2, 3, 4, 5, 6, 7, 8, 9], size=3)) # Three random numbers from the given array will be chosen

# To generate a random sample of values from a normally distributed with a specified mean and standard deviation
print(np.random.normal(loc=0, scale=1, size=10)) # 10 random values will be generated from a normally distributed data where 0 (loc) is the mean and 1 (scale) is the standard deviation

# To set a seed for a random number generation to ensure you get the same results later as well
print(np.random.seed(42)) # This ensures that any random numbers generated in the program are same every time you run the program. The seed value '42' can be any other value
```

AIM #2: Basic statistics
For the given dataset on sleep health and lifestyle, do the following
1. Using only pandas, load the dataset, calculate mean 'Sleep Duration', 'Systolic Blood Pressure', 'Diastolic Blood Pressure', 'Heart Rate' and 'Daily Steps'.
2. Do the same as in Step 1 using only NumPy
3. Using only pandas, first calculate correlation (across only the numerical variables), and then separate correlation between...
    Sleep duration and Age
    Sleep duration and Heart rate
    Sleep duration and Daily steps
4. Using only NumPy, do the same as Step 3
5. Using pandas only, calculate standard deviation for 'Sleep Duration'. 
6. Usiong NumPy only, calculate standard deviation for 'Sleep Duration'. 
7. Calculate the time difference between using pandas and NumPy, right from the step of loading the dataset to the final standard deviation step. 
    5.1. Which one is faster and why?

```python
import numpy as np
import time

# Start timing
start_time_numpy = time.time()

# Loading the CSV file as an array in NumPy
data = np.genfromtxt('sleep_health.csv', delimiter=',', skip_header=0, dtype=None, encoding='utf-8', names=True)

# Create separate arrays for the individual columns
person_id = data['Person_ID'].astype(int)
sleep_duration = data['Sleep_Duration'].astype(float)

# Calculate the correlation coefficient
correlation_matrix = np.corrcoef(person_id, sleep_duration)
correlation = correlation_matrix[0, 1]

# Store the result in a dictionary
sleep_duration_correlations_numpy = {
    'Persion_ID': correlation,
}

# Calculate mean using NumPy
# Assuming columns 1 through 5 are numeric and correctly indexed
mean_numpy = np.mean([data['Person_ID'], data['Sleep_Duration']])

# Calculate the correlation matrix for all numeric columns
numeric_columns = [data['Person_ID'], data['Sleep_Duration']]
correlation_numpy = np.corrcoef(numeric_columns)

# Calculate Standard Deviation for sleep_duration
std_numpy = np.std(sleep_duration)

# End timing
end_time_numpy = time.time()
numpy_duration = end_time_numpy - start_time_numpy

# Output results
print("Correlation:", sleep_duration_correlations_numpy)
print("Mean:", mean_numpy)
print("Correlation Matrix:\n", correlation_numpy)
print("Standard Deviation:", std_numpy)
print("Elapsed Time:", numpy_duration)
```

```python
# Write your code for AIM #3 here
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

df=pd.read_csv('sleep_health.csv')
sleep_duration = np.array(df['Sleep Duration'])
quality_of_sleep = np.array(df['Quality of Sleep'])
stress_level = np.array(df['Stress Level'])
physical_activity_level = np.array(df['Physical Activity Level'])
occupation = np.array(df['Occupation'])
bmi = np.array(df['BMI Category'])

# Bar plots

# Sleep Duration based on Quality of Sleep
unique_values, counts = np.unique(quality_of_sleep, return_counts=True)
plt.bar(unique_values, counts)
plt.title('Sleep Duration based on Quality of Sleep')
plt.show()

# Sleep Duration based on Stress Level
unique_values, counts = np.unique(stress_level, return_counts=True)
plt.bar(unique_values, counts)
plt.title('Sleep Duration based on Stress Level')
plt.show()

# Sleep Duration based on Physical Activity Level
unique_values, counts = np.unique(physical_activity_level, return_counts=True)
plt.bar(unique_values, counts)
plt.title('Sleep Duration based on Physical Activity Level')
plt.show()

# Sleep Duration based on Occupation
unique_values, counts = np.unique(occupation, return_counts=True)
plt.bar(unique_values, counts)
plt.title('Sleep Duration based on Occupation')
plt.show()

# Sleep Duration based on BMI Category
unique_values, counts = np.unique(bmi, return_counts=True)
plt.bar(unique_values, counts)
plt.title('Sleep Duration based on BMI Category')
plt.show()

# Box plots

# Sleep Duration vs Quality of Sleep
sleep_duration_list = [sleep_duration[quality_of_sleep == i] for i in np.unique(quality_of_sleep)]
plt.boxplot(sleep_duration_list)
plt.title('Sleep Duration vs Quality of Sleep')
plt.show()

# Sleep Duration vs Stress Level 
sleep_duration_list = [sleep_duration[stress_level == i] for i in np.unique(stress_level)]
plt.boxplot(sleep_duration_list)
plt.title('Sleep Duration vs Stress Level')
plt.show()

# Sleep Duration vs Physical Activity Level 
sleep_duration_list = [sleep_duration[physical_activity_level == i] for i in np.unique(physical_activity_level)]
plt.boxplot(sleep_duration_list)
plt.title('Sleep Duration vs Physical Activity Level')
plt.show()

# Sleep Duration vs Occupation
sleep_duration_list = [sleep_duration[occupation == i] for i in np.unique(occupation)]
plt.boxplot(sleep_duration_list)
plt.title('Sleep Duration vs Occupation')
plt.show()

# Sleep Duration vs BMI
sleep_duration_list = [sleep_duration[bmi == i] for i in np.unique(bmi)]
plt.boxplot(sleep_duration_list)
plt.title('Sleep Duration vs BMI')
plt.show()
```
